## Şubelere Göre Günlük Satış Miktarları (Haftanın Günleri)

### Subtask:
Şubelere göre haftanın günlerine göre günlük satış miktarlarını hesaplayın ve yoğun günleri tespit etmek için görselleştirin. Grafikte değerleri gösterin.

```python
# Convert 'Date' column to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Extract the day of the week
df['Day of Week'] = df['Date'].dt.day_name()

# Calculate the total quantity sold per branch and day of the week
daily_sales_by_branch = df.groupby(['Branch', 'Day of Week'])['Quantity'].sum().reset_index()

# Define the order of the days of the week for consistent plotting
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_sales_by_branch['Day of Week'] = pd.Categorical(daily_sales_by_branch['Day of Week'], categories=day_order, ordered=True)
daily_sales_by_branch = daily_sales_by_branch.sort_values('Day of Week')


# Create a grouped bar plot to visualize daily sales quantities by branch and day of the week
plt.figure(figsize=(12, 7))
ax = sns.barplot(x='Day of Week', y='Quantity', hue='Branch', data=daily_sales_by_branch)
plt.title('Şubelere Göre Günlük Satış Miktarları')
plt.xlabel('Haftanın Günü')
plt.ylabel('Toplam Satış Miktarı')

# Add values on top of the bars
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

## Şubelere Göre En Yoğun Saatler

### Subtask:
Her şube için günün hangi saatlerinin en yoğun olduğunu bulun ve görselleştirin.

```python
# Convert 'Time' column to datetime objects and extract the hour
df['Hour'] = pd.to_datetime(df['Time']).dt.hour

# Calculate the transaction count per branch and hour
hourly_sales_by_branch = df.groupby(['Branch', 'Hour']).size().reset_index(name='Count')

# Create a grouped bar plot to visualize hourly transaction counts by branch
plt.figure(figsize=(14, 7))
ax = sns.barplot(x='Hour', y='Count', hue='Branch', data=hourly_sales_by_branch)
plt.title('Şubelere Göre Saatlik Alışveriş Sayısı')
plt.xlabel('Saat')
plt.ylabel('Alışveriş Sayısı')
plt.xticks(range(24)) # Ensure all hours are shown on the x-axis

# Add values on top of the bars
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout() # Adjust layout to prevent labels from being cut off
plt.show()
```

## Şubelere Göre Günlük ve Saatlik Alışveriş Yoğunluğu

### Subtask:
Her şube ve haftanın her günü için saatlik alışveriş sayılarını hesaplayın ve çoklu grafikler ile görselleştirin.

```python
# Convert 'Time' column to datetime objects and extract the hour
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M').dt.hour

# Calculate the transaction count per branch, day of week, and hour
daily_hourly_sales_by_branch = df.groupby(['Branch', 'Day of Week', 'Hour']).size().reset_index(name='Count')

# Define the order of the days of the week for consistent plotting
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daily_hourly_sales_by_branch['Day of Week'] = pd.Categorical(daily_hourly_sales_by_branch['Day of Week'], categories=day_order, ordered=True)
daily_hourly_sales_by_branch = daily_hourly_sales_by_branch.sort_values(['Day of Week', 'Hour'])

# Create a FacetGrid to plot hourly sales for each day and branch
g = sns.FacetGrid(daily_hourly_sales_by_branch, col='Day of Week', row='Branch', sharey=True, height=4, aspect=1.2)
g.map_dataframe(sns.barplot, x='Hour', y='Count')

# Add grid to the plots
g.add_legend()
g.fig.subplots_adjust(top=0.9) # Adjust the top to make space for the title
g.fig.suptitle('Şubelere Göre Günlük ve Saatlik Alışveriş Yoğunluğu', y=1.02) # Add a main title


# Set titles and labels
g.set_titles("Şube: {row_name} | Gün: {col_name}")
g.set_axis_labels("Saat", "Alışveriş Sayısı")
g.set(xticks=range(0, 24, 2)) # Set x-axis ticks to show every other hour

for ax in g.axes.flat:
    ax.grid(True, linestyle='--', alpha=0.6)


plt.tight_layout()
plt.show()
```
